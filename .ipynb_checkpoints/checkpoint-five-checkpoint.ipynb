## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("df_daily.csv")
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
#for the purposes of this assignment, not selecting strongest correlation since some obvious ones exist. 
#exploring sedentary minutes and total minutes asleep
X = df['SedentaryMinutes']
Y = df['TotalMinutesAsleep']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
X_train
Y_train
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)

lr = sm.OLS(Y_train, X_train_sm).fit()

lr.params
lr.summary()
```

```python
plt.scatter(X_train, Y_train, label='Actual Data')  # Scatter plot of the data points
plt.plot(X_train, 716.78 + -0.4221*X_train, 'r', label='Regression Line')  # Plotting the regression line in red

# Add labels and title
plt.xlabel('Sedentary Minutes')
plt.ylabel('Total Minutes Asleep')
plt.title('Scatter Plot with Regression Line')

plt.legend()

plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
# Predicting y_value using traingn data of X
Y_train_pred = lr.predict(X_train_sm)

# Creating residuals from the y_train data and predicted y_data
res = (Y_train - Y_train_pred)

# Plotting the histogram using the residual values
fig = plt.figure()
sns.displot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('Y_train - Y_train_pred', fontsize = 15)
plt.show()
```
